```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how="left")
```

```python
final_df.head(2)
```

```python
## To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## Pie chart -Top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%",explode=[0.1,0.1,0.2])
```

Observation : Zomato maximun records/transactions are from India
After that USA and Then Unite Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

## Observation 
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When ratings are between 4.0 to 4.4 ---> very good
3. When ratings are between 3.5 to 3.9 ---> good
4. When ratings are between 2.5 to 3.4 ---> average
5. When ratings are between 1.8 to 2.4 ---> poor
6. When ratings are between 0.0 ---> Not rated

```python
ratings.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=(12,6)
sns.barplot(data=ratings,x="Aggregate rating",y="Rating count")
```

```python
sns.barplot(data=ratings,x="Aggregate rating",y="Rating count",hue="Rating color",palette=["blue","red","orange","yellow","green","green"])
```

## Observation 
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(data=ratings,x="Rating color",hue="Rating color",palette=["blue","red","orange","yellow","green","green"])
```

```python
# find the countries name that has given 0 rating
final_df[final_df["Rating color"]=="White"].groupby("Country").size().reset_index()
```

## Observation
India has given maximum number of 0 ratings
